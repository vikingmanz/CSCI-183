```python
#3

password = "That1guy@"

special_characters = ['!','@',"#",'$','&']
def valid_password(password):
    l = 0 
    u =0
    a = 0
    sc = 0
    if len(password) <=6 or len(password)>= 20:
        print("Invalid, Password length should be between 6 to 20 characters long")
    for c in password:
        if c.islower():
             l+=1
        if c.isupper():
             u+=1
        if c.isnumeric():
             a+=1
        if c in special_characters:
            sc+=1
    if l >= 1 and u >= 1 and a >= 1 and sc >= 1:
        print("Valid password")
    if l == 0:
        print("Invalid password, No lowercase letters include at least one")
    if u == 0:
        print("Invalid password, No uppercase letters include at least one")
    if a == 0:
        print("Invalid password, No numbers include at least one")
    if sc ==0:
        print("Invalid password! No special characters include at least one")
valid_password(password)
password = "That1guy@"
valid_password(password)
password = "THAT1guy"
valid_password(password)
password = "that1guy@"


```
